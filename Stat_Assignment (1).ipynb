# Sales Data

```python
data1=pd.read_csv('Sales_add.csv')
```

```python
data1.head()
```

```python
from scipy.stats import ttest_ind, chi2_contingency
```

a. If there is any increase in sales after digital marketing

```python
before_marketing = data1['Sales_before_digital_add(in $)']
after_marketing = data1['Sales_After_digital_add(in $)']
t_stat, p_value = ttest_ind(before_marketing, after_marketing)

if p_value < 0.05:
    print("There is a significant increase in sales after implementing digital marketing.")
else:
    print("There is no significant increase in sales after implementing digital marketing.")
```

b.Dependency between region and manager

```python
contingency_table = pd.crosstab(data1['Region'], data1['Manager'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
if p < 0.05:
    print("There is a significant dependency between Region and Manager.")
else:
    print("There is no significant dependency between Region and Manager.")
```
